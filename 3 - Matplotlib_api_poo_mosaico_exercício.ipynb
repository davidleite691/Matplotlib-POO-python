# Exercícios de mosaico de gráficos

Considere os dados da primeira célula para um aplicativo de uma pequena empresa de software. Faça um mosaico com 4 gráficos:

1. Gráfico de linhas com os novos usuários por mês
2. Gráfico de barras com a receita mensal
3. Gráfico de linhas com os usuários ativos diários
4. Gráfico de barras empilhadas com os tickets de suporte abertos e fechados

Os gráficos 1 e 2 devem ter uma linha própria no mosaico. Os gráficos 3 e 4 devem compartilhar a mesma linha.


```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
novos_usuarios = [100, 80, 110, 140, 150, 160, 170, 120, 110, 150, 200, 140]
usuarios_ativos_diarios = [1000, 750, 900, 1300, 1400, 1500, 1200, 1200, 1500, 1900, 2000, 1800]
receita_mensal = [10000, 8500, 9800, 10000, 11000, 12000, 10000, 10500, 13000, 13000, 20000, 17000]
tickets_suporte_abertos = [110, 75, 100, 140, 170, 160, 100, 130, 190, 200, 210, 140]
tickets_suporte_fechados = [100, 85, 100, 120, 130, 150, 140, 110, 130, 120, 150, 200]
```

```python
mosaico = "AA;BB;CD"
fig = plt.figure(figsize=(12,15))
espacamento = {"wspace":0.3, "hspace":0.5}
axs = fig.subplot_mosaic(mosaico, gridspec_kw=espacamento)

# 1 - Gráfico de linha
axs["A"].set_title("Novos usuários por mês")
axs["A"].plot(meses, novos_usuarios, marker = "o")

for i,num in enumerate(novos_usuarios):
    axs["A"].annotate(novos_usuarios[i], xy=(meses[i],novos_usuarios[i]), xytext=(-10, 5), textcoords='offset points')

axs["A"].set_frame_on(False)
axs["A"].set_yticks([])
axs["A"].set_ylabel("Quantidade de novos usuários")
axs["A"].tick_params(axis = "x", length = 0)

# 2 - Gráfico de barra
axs["B"].set_title("Receita mensal")
barras = axs["B"].bar(meses, receita_mensal)
axs["B"].bar_label(barras, fontweight = "bold")

axs["B"].set_yticks([])
axs["B"].set_ylabel("Receita mensal em R$")
axs["B"].set_frame_on(False)

# 3 - Gráfico de linha
axs["C"].set_title("Novos usuários por mês")
axs["C"].plot(meses, usuarios_ativos_diarios, marker = "o")

axs["C"].set_frame_on(False)
axs["C"].set_ylabel("Quantidade de novos usuários")
axs["C"].tick_params(axis = "x", length = 0)

# 4 - Gráfico de barras empilhadas
axs["D"].set_title("Receita mensal")
barras_suporte_aberos = axs["D"].bar(meses, tickets_suporte_abertos, label = "Abertos")
barras_suporte_fechados = axs["D"].bar(meses, tickets_suporte_fechados, bottom = tickets_suporte_abertos, label = "Fechados")
axs["D"].bar_label(barras_suporte_aberos, fontweight = "bold", fontsize = 8, padding=-12)
axs["D"].bar_label(barras_suporte_fechados, fontweight = "bold", fontsize = 8, padding=-12)

axs["D"].set_yticks([])
axs["D"].set_ylabel("Receita mensal em R$")
axs["D"].set_frame_on(False)
axs['D'].legend()

fig.suptitle('Dados do App X')
plt.show()
```
